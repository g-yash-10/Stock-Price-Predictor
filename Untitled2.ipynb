```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Downloading data from YahooFinance
import yfinance as yf
dataset = yf.download("AAPL", start="2017-01-01", end="2021-02-28")
dataset.to_csv('test.csv')
data=pd.read_csv('test.csv')
```

```python
data
```

```python
#Extracting useful data
df = pd.DataFrame(data, columns=['Date','Close'])
df
```

```python
#Splitting data in training and testing
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.20)

x_train = np.array(train.index).reshape(-1, 1)
y_train = train['Close']

x_test = np.array(test.index).reshape(-1, 1)
y_test = test['Close']
```

```python
fig, ax = plt.subplots()
ax.plot(df['Date'], df['Close'])
```

```python
#Fitting using Linear Regression using training set
from sklearn.linear_model import LinearRegression
rgr=LinearRegression()
rgr.fit(x_train,y_train)
```

```python
#Predicting the test dataset
y_pred=rgr.predict(x_test)
print(x_train)
```

```python
#Visualising the dataset
plt.title('Close Stock Price History [2017 - ]', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Closing Stock Price in $', fontsize=14)
plt.scatter(x_train, y_train, color='red')
plt.plot(x_train, rgr.predict(x_train), color='blue')
plt.show()
```
